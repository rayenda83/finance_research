## Load the library

```python
import yfinance as yf
import pandas as pd
from datetime import datetime
```

```python
stock = 'BBCA.JK'

today = datetime.today()    # Tanggal hari ini
start = datetime(2023,1,1)  # Tanggal mulai data

# Menggunakan .download method utk lakukan penarikan data
price = yf.download(stock, start, today)
```

```python
price
```

```python
price.to_csv('bbca.csv', index = True)
```

```python
price.reset_index(drop = False)
```

```python
stocks = ['FILM.JK', 'GOTO.JK', 'PTBA.JK', 'INDY.JK', 'BUKA.JK']
```

```python
for stock in stocks:
    price = yf.download(stock, start, today)
    price.to_csv(f'{stock}.csv', index = True)
```
